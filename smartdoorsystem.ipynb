```python
# ===============================
#   Smart Door Access System
#         UMaT - ORIC
# ===============================

class Staff:
    def __init__(self, s_name, access_code):
        self.s_name = s_name
        self.__access_code = access_code   # Private attribute

    # VIEW access code (authorized only)
    @property
    def access_code(self):
        return self.__access_code

    # UPDATE access code with length validation
    @access_code.setter
    def access_code(self, new_code):
        if len(new_code) >= 4:
            self.__access_code = new_code
            print("✅ Access code updated successfully!")
        else:
            print("❌ Code too short! Must be at least 4 characters.")

    # DISPLAY staff information
    def display_info(self):
        print(f"Staff Name  : {self.s_name}")
        print(f"Access Code : {self.__access_code}")


# ===============================
#         Testing It
# ===============================

# Create staff object
staff1 = Staff("Ekow Daniels", "1234")

# Display info
print("=" * 35)
print("       STAFF INFORMATION")
print("=" * 35)
staff1.display_info()

# View access code using property
print(f"\nCurrent Code: {staff1.access_code}")

# Update with valid code
print("\nUpdating access code to 'Ek2007!'...")
staff1.access_code = "Ek2007!"

# Update with invalid code
print("Updating access code to '12'...")
staff1.access_code = "12"

# Display updated info
print()
print("=" * 35)
print("      UPDATED INFORMATION")
print("=" * 35)
staff1.display_info()
```
